## Data Cleaning and Preprocessing  

```python
#check for null values
data.isnull().sum()
```

```python
data.info()
```

We focused on post-1950 data for our comparative analysis and global EDA since some countries are missing significant data before that time. Major carbon emitting countries' emission activites intensified post 1950 period, especially India. For our modelling process, however, focusing on India, the data form 1878 is continous, so we include all data for India from 1900. First, We'll be predicting and Forecasting the per capita CO2 emissions for India

```python
#only include India co2 per capita data post-1900
df_India = data
df_India = df_India[df_India['year'] >= 1900]
df_India = df_India[df_India['country'] == 'India']
df_India = df_India[['year', 'co2_per_capita']]

#observe shape of reformatted df
print(df_India.shape)
```

**Feature Scaling**

```python
scaler = MinMaxScaler(feature_range=(0,1))
training_scaled_data = scaler.fit_transform(df_India)
training_scaled_data.shape

#observe scaled data
#training_scaled_data
```

```python
# Fix the random number seed  so that our results are reproducible.
np.random.seed()
```

**Train-Test Split**

```python
# Data from 1900-2020: We will split the data into 80% training set and 20% validation set.
training_set = int(len(df_India)* 0.8)
training_set
```

```python
# Filter the scaled train dataset
train_data = training_scaled_data[0:training_set , : ]
train_data.shape
```

```python
# Filter the scaled test dataset
test_data = training_scaled_data[training_set: , : ]
test_data.shape
```

**Creating Sequences**

```python
# Train the dataset in the cycle of 4 years, so 1960-1963; 1961-1964 and so on.
X_train = []
y_train = []
memory_range = 4
for i in range(memory_range, len(train_data)):
    X_train.append(train_data[i-memory_range:i, 0])
    y_train.append(train_data[i, 0])

    
# Converting into numpy sequences to train the LSTM model
X_train, y_train = np.array(X_train), np.array(y_train) 
```

```python
# Reshape the training data to 3 dimensional for LSTM input.
X_train = X_train.reshape(X_train.shape[0],X_train.shape[1] , 1)
```
